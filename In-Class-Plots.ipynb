```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Imported dataset from Kaggle 
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

## Histograms 

```python
plt.figure(figsize=(8,5))
plt.hist(df['math score'], color='m', bins = 30)
plt.ylabel('Frequency')
plt.title('Math Score Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['reading score'], color='r', bins = 30)
plt.ylabel('Frequency')
plt.title('Reading Score Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['math score'], color='b', bins = 30)
plt.ylabel('Frequency')
plt.title('Writing Score Distribution')
plt.show()
```

## Bar Graph 

```python
x = df['race/ethnicity']
y = df['writing score']

plt.title('Writing score for groups')
plt.bar(x,y)
plt.show()
```

```python
x = df['race/ethnicity']
y = df['math score']

plt.title('Math score for groups')
plt.bar(x,y)
plt.show()
```

```python
x = df['race/ethnicity']
y = df['reading score']

plt.title('Reading score for groups')
plt.bar(x,y)
plt.show()
```

## Grouped Bar Graph 

```python
plt.rcParams['figure.figsize'] = (10, 8)
x = pd.crosstab(df['gender'], df['race/ethnicity'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = False)
plt.title('Gender vs Race', fontweight = 30, fontsize = 20)
plt.show()
```

## Pie Charts 

```python
# creating a new column pass_math, which will tell us whether the students have passed or failed
passing_marks = 50 
df['pass_math'] = np.where(df['math score']< passing_marks, 'Fail', 'Pass')
df['pass_math'].value_counts().plot.pie(colors = ['purple', 'red'])

plt.title('Pass/Fail in Maths', fontweight = 30, fontsize = 20)
plt.xlabel('result')
plt.ylabel('count')
plt.show()
```

```python
# creating a new column pass_reading, this column will tell us whether the students are pass or fail
passing_marks = 50 
df['pass_reading'] = np.where(df['reading score']< passing_marks, 'Fail', 'Pass')
df['pass_reading'].value_counts().plot.pie(colors = ['lightgreen', 'red'])

plt.title('Pass/Fail in Reading', fontweight = 30, fontsize = 20)
plt.xlabel('result')
plt.ylabel('count')
plt.show()
```

```python
# creating a new column pass_writing, this column will tell us whether the students are pass or fail
passing_marks = 50 
df['pass_writing'] = np.where(df['writing score'] < passing_marks, 'Fail', 'Pass')
df['pass_writing'].value_counts().plot.pie(colors = ['orange', 'red'])

plt.title('Pass/Fail in Writing', fontweight = 30, fontsize = 20)
plt.xlabel('result')
plt.ylabel('count')
plt.show()
```

## Scatter Plot 

```python
x = df['math score']
y = df['writing score']
fig=plt.figure()
plt.scatter(x, y, color='r')
plt.xlabel("Math Scores")
plt.ylabel("Writing Scores")
plt.title("Math Scores vs Writing Scores")
plt.show()
```
